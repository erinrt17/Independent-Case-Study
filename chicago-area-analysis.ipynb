# Data Selection
The dataset I'm using is from the Chicago Metropolitan Agency for Planning. It has demographic information for 77 areas in Chicago. 

```python
import pandas as pd
import numpy as np
import plotly
import plotly.express as px
import plotly.graph_objects as go
```

Import Dataset

```python
df = pd.read_csv('https://github.com/erinrt17/Chicago-Neighborhood-Independent-Case-Study/raw/main/ReferenceCCAProfiles20142018.csv')
```

Check which columns are empty

```python
df.isnull()
```

Drop empty columns

```python
for column in df:
  if df[column].isnull().sum() >5:
    print('{0} has {1} null values'.format(column, df[column].isnull().sum()))
    df.drop(column, axis=1, inplace=True)
```

Check dataset to make sure that empty columns were dropped

```python
df.shape
```

Sort dataset by my age group 


```python
df.sort_values('A20_34', ascending=False)
```

Sort data by median income

```python
df.sort_values('MEDINC', ascending=False)
```

Sort data by my salary range

```python
df.sort_values('INC_50_75K', ascending=False)
```

Group by residents' employment type to determine which is assocaited with the highest median income


```python
df_employment = df.groupby('RES_NAICS1_TYPE', as_index=False)
df_employment.agg({'MEDINC': 'mean'})
```

Determine Chicago's distribution of neighborhoods with a bachelor's degree or higher education

```python
fig = px.box(df, y='BACH', title='Bachelor Degree Distribution')
fig.show()
```

Determine which neighborhoods have access to open space

```python
fig = px.density_heatmap(df,x='OPEN_SPACE_PER_1000', y='GEOG', title='Two Bedroom Apartments with Access to Open Space')
fig.show()
```

Distrubtion of Median Income by my Age Group 

```python
fig = px.scatter(df, x='A20_34', y='MEDINC', title='Median Income by my Age Group')
fig.show()
```

Areas with population in my age group and salary range


```python
fig = px.scatter_3d(df, x='A20_34', y='INC_50_75K', z='GEOG', title='Areas with my age and salary range with two bedroom apartments')
fig.show()
```

Pie chart of neighborhoods with population in my age group

```python
fig=px.pie(df, values='A20_34', names='GEOG', title='My age group by neighborhood')
fig.show()
```

After analyzing the data in python, I have a relative basis for neighborhoods that I would like to live in.
